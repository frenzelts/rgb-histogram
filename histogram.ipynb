## Histogram for RGB Image

```python
import cv2 
import os
from matplotlib import pyplot as plt 
import cv2 
import numpy as np
from glob import glob
```

#### Define image path

```python
filepath = os.path.join('.', 'images', 'ISIC_0024306.JPG')
filename = os.path.basename(filepath).split('.')[0]
```

#### Read image

```python
im = cv2.imread(filepath, cv2.IMREAD_UNCHANGED)
```

#### Create new figure and color enum

```python
fig = plt.figure()
color = ('b','g','r')
```

#### Calculate pixel distribution using calcHist then plot to histogram

```python
for i,col in enumerate(color):
    histr = cv2.calcHist([im],[i],None,[256],[0,256])
    sorted = histr[np.argsort(histr[:, 0])]
    plt.plot(histr,color = col)
plt.xlim([0,256])
plt.xlabel('Pixel values')
plt.ylabel('Number of pixels')

save_dir = os.path.join('.', 'histogram')
plt.savefig(os.path.join(save_dir, filename+'.png'))

plt.show()
```

### Generate histogram using loop for specific directory

```python
images_dir = os.path.join('.', 'images')

for x in glob(os.path.join(images_dir, '*.jpg')):
    filepath = os.path.abspath(x)
    filename = os.path.basename(filepath).split('.')[0]
    
    im = cv2.imread(filepath, cv2.IMREAD_UNCHANGED)
    
    fig = plt.figure()
    color = ('b','g','r')
    
    for i,col in enumerate(color):
        histr = cv2.calcHist([im],[i],None,[256],[0,256])
        sorted = histr[np.argsort(histr[:, 0])]
        plt.plot(histr,color = col)
    plt.title('Histogram '+filename)
    plt.xlim([0,256])
    plt.xlabel('Pixel values')
    plt.ylabel('Number of pixels')
    save_dir = os.path.join('.', 'histogram')
    plt.savefig(os.path.join(save_dir, filename+'.png'))
    plt.show()
```
